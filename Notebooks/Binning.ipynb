```python
import pandas as pd
import numpy as np
```

```python
col1 = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']
col2 = [25, 35, 45, 30, 50, 60, 70, 80, 90, 100, 10]

df = pd.DataFrame({"Names": col1, "Marks": col2})
df
```

```python
bins = [0, 25, 50, 75, 100]
groups = ['Fail', 'Average', 'Good', 'Brilliant']
```

```python
df['Grades'] =  pd.cut(df.Marks, bins, labels = groups)
df
```
