```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
file_path = r'D:\DataScientist\Andrew NG\Data Visualization Python\Week3\Assignment\Topic_Survey_Assignment.csv'
# importing Data set
df = pd.read_csv(file_path, index_col=0)
```

```python
# let's examine the data set data
df
```

```python
# sirting data frame with 'Very interested' by desc
df_sorted = df.sort_values(['Very interested'], ascending=[False])
df_sorted
```

```python
#converting numbers to percentage
total_survey_res = 2233
precision = 2
df_sorted['Very interested']=round((df_sorted['Very interested'] / total_survey_res)*100, precision)
df_sorted['Somewhat interested']=round((df_sorted['Somewhat interested'] / total_survey_res)*100, precision)
df_sorted['Not interested']=round((df_sorted['Not interested'] / total_survey_res)*100, precision)
df_sorted
```

```python
bar_plot = df_sorted.plot(kind='bar',
            figsize=(20, 8),
            width=0.8,
            color=['#5cb85c', '#5bc0de', '#d9534f'],
           )

bar_plot.spines['top'].set_visible(False)
bar_plot.spines['left'].set_visible(False)
bar_plot.spines['right'].set_visible(False)
bar_plot.axes.get_yaxis().set_visible(False)
bar_plot.tick_params(labelsize=14)
bar_plot.legend(fontsize=14)

bar_plot.set_title("Percentage of Respondents' Interest in Data Science Area", fontsize=16)
for p in bar_plot.patches:
    bar_plot.annotate(np.round(p.get_height(),decimals=2),
                (p.get_x()+p.get_width()/2., p.get_height()),
                ha='center',
                va='center',
                xytext=(0, 10),
                textcoords='offset points',
                fontsize = 14
               )
plt.show()
```

```python
ax = df_sorted.plot(kind='bar', figsize=(20, 8),width=0.8, color=['#5cb85c', '#5bc0de', '#d9534f'])

ax.set_title('Percentage of Respondents Interest in Data Science Area',fontsize=16)
ax.set_ylabel('Number')
ax.set_xlabel('Topics')
ax.legend(fontsize = 14)
ax.set_frame_on(False)
ax.set_xlabel(False).set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.set_xticklabels(('Data Analysis / Statistics','Machine Learning','Data Visualization','Big Data (Spark / Hadoop)','Deep Learning','Data Journalism'))
```
