```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y1 = np.random.randn(70)
y2 = np.random.randn(70)

# y1 and y2 should be same size

plt.scatter(y1, y2)
plt.show()
```

# adjusting axis values

```python
y1 = np.random.randn(70)
y2 = np.random.randn(70)

# y1 and y2 should be same size

plt.scatter(y1, y2)
plt.axis([-3, 3, -3, 3]) #plt.axis([Xmin, Xmax, Ymin, Ymax])
plt.show()
```

# common x asix, and different markers

```python
x = np.random.randn(70)

y1 = np.random.randn(70)
y2 = np.random.randn(70)

plt.scatter(x, y1, marker="o", label="x with y1")
plt.scatter(x, y2, marker="v", label="x with y2")
plt.axis([-3,3, -3, 3])
#title
plt.legend("upper right")

plt.show()
```

```python
# color 
x = np.random.randn(70)

y1 = np.random.randn(70)
y2 = np.random.randn(70)

plt.scatter(x, y1, marker="o", label="x with y1", c='r')
plt.scatter(x, y2, marker="v", label="x with y2", c='g')
plt.axis([-3,3, -3, 3])
#title
plt.legend("upper right")

plt.show()
```
