### 2. Label Encoder

```python
col1 = ['A', 'B','C', 'D', 'E', 'F']
col2 = ['Male', 'Female','Female', 'Male', 'Male', 'Female']
col3 = [75, 89, 88, 98, 100, 99]
col4 = ['Street 1', 'Street 2', 'Street 1', 'Street 3', 'Street 3', 'Street 2']

df1 = pd.DataFrame({"Name": col1, "Gender": col2, "Marks": col3, "Address": col4})
df1
```

```python
df1.info()
```

```python
cat_cols = [f for f in df1.columns if df1[f].dtype == 'object']
cat_cols
```

```python
label_encoder = LabelEncoder()
df1 = df1[cat_cols].apply(label_encoder.fit_transform)
df1
```
